Before we discuss optimizers, let me bring up the Loss Function, used for accuracy of the dataset calcuations. It is defined by:
${-(ylog(p) + (1-y)log(1-p))}$.

Optimizers are literally what the name is; they optimize based off the results of the loss function to increase accuracy among the data. The loss function acts as a trial and error and the optimizer is used as a "clean up".
Stochastic Gradient Descent is an optimizer that gets a guess (trial and error) on the errors and then follows with new weights (clean up).
